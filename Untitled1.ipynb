```python
import pandas as pd
import numpy as np

# Creating synthetic network traffic data
data = {
    'duration': np.random.rand(100),
    'protocol_type': np.random.choice(['tcp', 'udp', 'icmp'], 100),
    'service': np.random.choice(['http', 'smtp', 'ftp'], 100),
    'src_bytes': np.random.randint(100, 5000, 100),
    'dst_bytes': np.random.randint(100, 5000, 100),
    'label': np.random.choice(['normal', 'attack'], 100) # The target we want to detect
}

df = pd.DataFrame(data)

# Saving to a file
df.to_csv('network_data.csv', index=False)
print("File 'network_data.csv' created successfully!")
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# 1. Convert text columns (like protocol, service, label) to numbers
le = LabelEncoder()
df['protocol_type'] = le.fit_transform(df['protocol_type'])
df['service'] = le.fit_transform(df['service'])
df['label'] = le.fit_transform(df['label'])

# 2. Separate data into inputs (X) and target (y)
X = df.drop('label', axis=1)
y = df['label']

# 3. Train the Model
model = DecisionTreeClassifier()
model.fit(X, y)

print("Model trained successfully! The AI is now ready to detect attacks.")
```

```python
# Check how well the model learned (Accuracy)
accuracy = model.score(X, y)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Test with a new (simulated) network connection
# [duration, protocol_type, service, src_bytes, dst_bytes]
new_connection = [[0.5, 1, 1, 1000, 500]]
prediction = model.predict(new_connection)

if prediction[0] == 1:
    print("Prediction: Normal Traffic")
else:
    print("Prediction: Attack Detected!")
```

```python
# Create a DataFrame with column names for the test input
new_data = pd.DataFrame([[0.5, 1, 1, 1000, 500]], 
                        columns=['duration', 'protocol_type', 'service', 'src_bytes', 'dst_bytes'])

# Predict using the named columns
prediction = model.predict(new_data)

if prediction[0] == 1:
    print("Prediction: Normal Traffic")
else:
    print("Prediction: Attack Detected!")
```
